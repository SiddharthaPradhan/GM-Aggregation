# Event log data

```python
# load eventLogs
event_df = pd.read_json(data_folder + event_file)
print(event_df.columns)
print(event_df['eventType'].unique())
event_df.head()
```

```python
# cols of interest for aggregation
event_cols_to_keep = ['studentRef', 'taskNumber',
                      'attemptNumber', 'eventType', 'timestamp', 'mistake', 'stars', 'attemptHlc']
# user interaction events: math actions, mistake events or hint events
action_events = ['mathStep', 'mathMistake', 'hideHint', 'showHint']
event_df = event_df[event_cols_to_keep]
```

```python
event_df = event_df.sort_values(
    by=['studentRef', 'taskNumber', 'attemptNumber', 'timestamp'])
```

```python
event_df.loc[:, 'previous_timestamp'] = event_df.groupby(
    ['studentRef', 'taskNumber'])['timestamp'].shift(1)
event_df['delta_time'] = event_df['timestamp'] - event_df['previous_timestamp']
event_df['delta_time_tuple'] = list(
    zip(event_df['delta_time'], event_df['eventType'].isin(action_events)))
```

```python
# remove redundant eventTypes: 'visitGameScreen'
event_df = event_df.loc[~event_df['eventType'].isin(['visitGameScreen'])]
```

```python
# TODO next time use attemptHlc to group
event_grouped = event_df.groupby(['studentRef', 'taskNumber', 'attemptNumber'])
```

```python
def get_pause_time(col: pd.Series) -> float:
    col = col.apply(pd.Series) # expand tuples
    action_times = col.loc[col[1] == True, 0]
    if len(action_times) == 0:
        return None
    else:
        return action_times.iloc[0].total_seconds()

def get_validity_first_step(col:pd.Series) -> bool:
    is_mistake = col == 'mathMistake'

    is_math_step = col == 'mathStep'
    if not is_math_step.any(): # if no math steps
        return None
    elif not is_mistake.any():  # if no math mistakes and at least one math step
        return True
    else:
        first_mistake_index = is_mistake.idxmax()
        first_math_step_index = is_math_step.idxmax()
        return first_math_step_index < first_mistake_index
    
def get_replay_attempt(completed_dur_attempt:pd.Series) -> pd.Series:
    if not completed_dur_attempt.any():
        return pd.Series([False]*len(completed_dur_attempt))
    else:
        first_completed_index = completed_dur_attempt.idxmax()
        return completed_dur_attempt.index > first_completed_index
```

```python
# stage one: info from all event types
attempt_agg_df = event_grouped.agg(
    start_time=('timestamp', 'first'),
    # end time needs to ignore time after solvingTask (e.g. reset)
    end_time=('timestamp', 'last'), # need to account for time after solving task and waiting before next attempt or moving on. 
    total_time = ('timestamp', lambda x: (x.iloc[-1] - x.iloc[0]).total_seconds()),
    pause_time=('delta_time_tuple', get_pause_time),
    num_steps = ('eventType', lambda x: (x == 'mathStep').sum()),
    num_errors=('eventType', lambda x: (x == 'mathMistake').sum()),
    num_keypad_errors=('mistake', lambda x: (x == 'keypad').sum()),
    num_shaking_errors=('mistake', lambda x: (x == 'tap').sum()),
    num_snapping_errors=('mistake', lambda x: (x == 'drag').sum()),
    num_hints = ('eventType', lambda x: (x == 'showHint').sum()),
    completed_dur_attempt=('eventType', lambda x: (x == 'solvedTask').any()),
    stars=('stars', 'max'),
    num_visits=('eventType', lambda x: (x == 'visitTask').sum()),
    validity_first_step = ('eventType', get_validity_first_step),
    attemptHlc=('attemptHlc', lambda x: x.mode().iloc[0]),
).reset_index()
attempt_agg_df['pause_time_ratio'] = attempt_agg_df['pause_time'] / attempt_agg_df['total_time']
attempt_agg_df['replay_attempt'] = attempt_agg_df.groupby(['studentRef', 'taskNumber'])[
    'completed_dur_attempt'].transform(get_replay_attempt)
```

# Check attempt and aggregated attempt merge

```python
attempt_merge_df = pd.merge(attempt_agg_df, attempt_df[attempt_df.columns.difference(['studentRef'])],
                            how='inner',
                            left_on='attemptHlc', right_on='clientCreatedAtHlc')
attempt_merge_df.drop(columns=['clientCreatedAtHlc', 'taskWork', 'taskWorkNumber'], inplace=True)
```
