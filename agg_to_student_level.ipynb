```python
import pandas as pd
```

```python
attempt_processed_df = pd.read_csv('./output/agg_attempt_merged_df.csv')
```

```python
attempt_processed_df['stars'].unique()
```

```python
attempt_processed_df['stars_1'] = attempt_processed_df['stars'] == 1
attempt_processed_df['stars_2'] = attempt_processed_df['stars'] == 2
attempt_processed_df['stars_3'] = attempt_processed_df['stars'] == 3
attempt_processed_df['stars_NA'] = attempt_processed_df['stars'].isna()
```

```python
student_problem_df = attempt_processed_df.groupby(['studentRef', 'taskNumber']).agg(
        attempts=('attemptNumber', 'max'),
        total_replays=('replay_attempt', 'sum'),
        completed_attempts=('completed_dur_attempt', 'sum'),
        num_errors=('num_errors', 'sum'),
        num_hints=('num_hints', 'sum'),
        total_time_spent=('timeSpent', 'sum'),
        total_three_star=('stars_3', 'sum'),
        total_two_star=('stars_2', 'sum'),
        total_one_star=('stars_1', 'sum'),
        total_incomplete=('stars_NA', 'sum'),
        avg_stars=('stars', 'mean')
).reset_index()
student_problem_df['completed_attempts'] = student_problem_df['completed_attempts'].fillna(0).astype(int)
student_problem_df['total_replays'] = student_problem_df['total_replays'].fillna(
    0).astype(int)
```

```python
student_problem_df.to_csv('./output/student_problem_level.csv', index=False)
```

```python
student_df  = student_problem_df.groupby('studentRef').agg(
    total_problems=('taskNumber', 'nunique'),
    total_attempts=('attempts', 'sum'),
    total_replays=('total_replays', 'sum'),
    total_completed_attempts=('completed_attempts', 'sum'),
    total_errors=('num_errors', 'sum'),
    total_hints=('num_hints', 'sum'),
    total_time_spent=('total_time_spent', 'sum'),
    total_one_star=('total_one_star', 'sum'),
    total_two_star=('total_two_star', 'sum'),
    total_three_star=('total_three_star', 'sum'),
    total_incomplete=('total_incomplete', 'sum'),
    avg_stars=('avg_stars', 'mean')
).reset_index()
    
```
